##<b>Question-3 : </b> 
#####Find Precision, Recall and F1 score in all cases.
<br>

Precision, recall, and F1 score are metrics commonly used to evaluate classification models.
<br>
*   Precision measures how many of the predicted positive instances are actually positive. It can be calculated as the ratio of true positives to the sum of true positives and false positives.

*   Recall (also called sensitivity) measures how many of the actual positive instances are correctly predicted. It can be calculated as the ratio of true positives to the sum of true 
  positives and false negatives.


*   F1 score combines precision and recall into a single metric that balances both. It is the harmonic mean of precision and recall and provides a single value that represents both measures.

##<b>Question-5 : </b>
#####Try any other techniques/algorithms in your research that could improve the accuracy.
<br>
We have chosen to implement a CNN-based model (Convolutional Neural Network) to improve upon the accuracy of detection.
<br>
<br>
The CNN-based model implemented, is designed to achieve high accuracy for image classification tasks, and has been built using a CNN model using Keras with TensorFlow backend. The model architecture follows a classic convolutional neural network structure that has proven to be effective in various image classification problems.
<br>
<br>
The model consists of multiple convolutional layers followed by max-pooling layers. Convolutional layers use filters to extract relevant features from the input images, and the max-pooling layers downsample the feature maps to capture the most important information. This combination of convolution and pooling layers allows the model to learn hierarchical representations of the input images.
<br>
<br>
The architecture also includes fully connected (dense) layers that are responsible for making final predictions. These dense layers receive the flattened output of the previous convolutional layers and employ non-linear activation functions to capture complex relationships between the learned features.
<br>
To ensure the model's ability to generalize and prevent overfitting, regularization techniques like Dropout are incorporated. Dropout randomly deactivates a specified percentage of neurons during training, forcing the network to learn robust representations.
<br>
<br>
Overall, this CNN-based model provides a strong foundation for achieving high accuracy in image classification tasks. It leverages the power of deep learning and convolutional neural networks to automatically learn relevant features from the input images, leading to improved accuracy compared to traditional machine learning algorithms.

PREPROCESSING IMAGES 

```python
def imgLoader(filepath):
  img = cv2.imread(filepath)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  return gray

#preprocessing image- thresholding, dilation, erosion
def preprocess(img):
  threshold = threshold_otsu(img)
  img_binary = img > threshold
  kernel = np.ones((3,3), np.uint8)
  img_dilated = dilation(img_binary, kernel)
  img_eroded = erosion(img_dilated)
  return img_eroded

#store letter names
def df_colNames(df):
  rows, cols=df.shape
  column_names = []
  [column_names.append("pixel"+str(x)) for x in range(cols-1)]
  column_names.append("label")
  df.columns=column_names
  return df

#function to take out accuracy score
def acc(y_true : np.ndarray, y_pred : np.ndarray) -> float:
    return round(accuracy_score(y_true, y_pred) * 100, 2)
```

FUNCTION TO STORE IMAGES IN A DATAFRAME

```python
def datagen(train_dir):
  trainDir=[f for f in os.listdir(train_dir) if not f.startswith('.')]
  df = pd.DataFrame()
  for j, filename in enumerate(trainDir):
    img=imgLoader(train_dir+filename)
    filename=filename.replace('.png', '')
    filename_list=list(filename)
    img=preprocess(img)
    img=img.astype(int)
    tiles=len(filename_list)
    imgheight, imgwidth = img.shape[0], img.shape[1] 
    for i in range(0, imgwidth, imgwidth//tiles):
      i1=i+imgwidth//tiles
      tmp=img[:,i:i1]
      tmp=tmp.flatten()
      tmp=np.append(tmp, filename_list[i//(imgwidth//tiles)])
      df=df.append(pd.DataFrame(tmp.reshape(1, -1)))
  df.reset_index(inplace=True,drop=True)
  df=df_colNames(df)
  print(df.shape)
  return df
```

FUNCTION FOR EVALUATION METRICS OF THE REQUIRED MODEL

```python
def getEvaluationMetrices(df, model_name, test_size=0.25, seed=40):
  features = [col for col in df.columns if col.startswith('pixel')]
  df[features]=df[features].astype(float)
  X_train, X_val, y_train, y_val = train_test_split(df[features], df['label'], test_size=0.2)

  if model_name == "LR":
    model = LogisticRegression(fit_intercept=True,multi_class='auto',penalty='l2')
  
  elif model_name == "DT":
    model = DecisionTreeClassifier(max_depth=10, random_state=seed)

  elif model_name == "RF":
    model = RandomForestClassifier(n_estimators=1000, random_state=seed)

  else :
    model = SVC(kernel='rbf')
  
  model.fit(X_train, y_train)
  train_preds_baseline = model.predict(X_train)
  val_preds_baseline = model.predict(X_val)
  acc_baseline_train = acc(train_preds_baseline, y_train)
  acc_baseline_val = acc(val_preds_baseline, y_val)
  print(f'Training accuracy for our baseline (using all pixel features): {acc_baseline_train}%')
  print(f'Validation accuracy for our baseline (using all pixel features): {acc_baseline_val}%')
  print()
  print(confusion_matrix(y_true=y_val, y_pred=val_preds_baseline))
  print()
  print(classification_report(y_true=y_val, y_pred=val_preds_baseline))
  return 
```

APPLYING CNN TO IMPROVE ACCURACY

```python
train=df
```

```python
Y_train = train["label"]
# Drop 'label' column
X_train = train.drop('label',axis = 1) 
```

```python
# Reshape image in 3 dimensions 
X_train = X_train.values.reshape(-1,100,20,1)

Y_train=Y_train.apply(lambda x: ord(x)-97)

# Encoding labels 
Y_train = to_categorical(Y_train, num_classes = 26)

# Split the train and the testing set for the fitting
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.1, random_state=2)
X_train=X_train.astype(float)
X_val=X_val.astype(float)
```

```python
# Set the CNN model 
# my CNN architechture is In -> [[Conv2D->relu]*2 -> MaxPool2D -> Dropout]*2 -> Flatten -> Dense -> Dropout -> Out

model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu', input_shape = (100,20,1)))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(26, activation = "softmax"))
```

```python
#Defining the optimizer
optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
```

```python
# Compiling the model
model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
# Setting a learning rate annealer
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', 
                                            patience=3, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.00001)
```

```python
epochs = 19
batch_size = 86
```

```python

history = model.fit(X_train, Y_train, batch_size = batch_size, epochs = epochs, 
        validation_data = (X_val, Y_val), verbose = 2)
```

PLOTTING CONFUSION MATRIX FOR CNN

```python
def plot_confusion_matrix(cm, classes,
                          title='Confusion Matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Predict the values from the validation dataset
Y_pred = model.predict(X_val)

# Convert predictions classes to one hot vectors 
Y_pred_classes = np.argmax(Y_pred,axis = 1) 

# Convert validation observations to one hot vectors
Y_true = np.argmax(Y_val,axis = 1) 

# compute the confusion matrix
confusion_mtx = confusion_matrix(Y_true, Y_pred_classes) 

# plot the confusion matrix
plot_confusion_matrix(confusion_mtx, classes = range(26)) 

#0-a,1-b,2-c...........25-z
```
